```
# ✅ Sprint 1a – Python Əsas Sintaksis və Operatorlar
a = str(123)
print(a, type(a))

b = int(19.99)
print(b)

c = int("500")
print(c / 2)

a = 8
b = 12
print(a < b)
print(a == b)

x = 5
y = 10
z = 15
print((x < y) and (y < z))

print(25 // 4)  # Tam bölmə
print(25 % 4)   # Qalıq
print(25 / 4)   # Adi bölmə

print(3 ** 4)

qiymet = 75.5
qiymet_int = int(qiymet)
print(qiymet_int, type(qiymet_int))

n = 20
print((n > 10) or (n < 5))
print((n > 15) and (n < 25))

x = float("42.8")
print(x, type(x))
y = int(x)
print(y, type(y))
```

```
# ✅ Sprint 1b – Mətnlərlə İş
s = "Programming"
print(len(s))
print(s[0])

s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)

text = "Python"
print(text[-1])

s = "Artificial"
print(s[:3])

word = "Code"
print(word[::-1])

s = "abcdefgh"
print(s[::2])

text = "data"
print(text.upper(), text.lower())

s = "Python-R-Java"
print(s.split("-"))

ad = "Ayxan"
yas = 25
print(f"{ad} {yas} yaşındadır")

s = "salam-dunya"
print(s.replace("-", " "))
```

```
# ✅ Sprint 2a – Listlərlə İş
reqemler = [5, 10, 15, 20]
print(len(reqemler))

reqemler.append(25)
print(reqemler)

reqemler.insert(2, 12)
print(reqemler)

list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2)

print(reqemler[2:4])

reqemler[0] = 50
print(reqemler)

print(19 in reqemler)

l = ["a", "b", "a", "c"]
print(l.count("a"))

l = ["x", "y", "x", "z"]
while "x" in l:
    l.remove("x")
print(l)

nums = [7, 2, 9, 1]
nums.sort(reverse=True)
print(nums)

print([x for x in reqemler if x > 10])
```

```
# ✅ Sprint 2b – Pandas və Series, DataFrame
import pandas as pd

s1 = pd.Series([10, 20, 30, 40])
s1.index = ['w', 'x', 'y', 'z']
print(s1)

s2 = pd.Series({'p': 5, 'q': 10, 'r': 15})
print(s2)

print(s2['q'])
print(s1[s1 > 25])
print(s1[s1 > 20] / 10)

df1 = pd.DataFrame([[1, 2], [3, 4]])
df1.index = ['r1', 'r2']
df1.columns = ['c1', 'c2']
print(df1)

df2 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print(df2)

print(df2[df2['A'] > 1])
```
